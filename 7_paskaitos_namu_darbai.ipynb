```python
# 20. Sukurkite funkciją is_tropical() winemag-data.csv duomenų rinkinyje,
# kuri priima string kintamąjį ir grąžina ar jame yra žodis 'tropical' (True/False)
```

```python
def is_tropical(text):
    x = False
    if 'tropical' in text.lower():
        x = True
    else:
        x = False
    return x 
```

```python
# 21. Naudodami .map() metodą ir nežymiai modifikuotą ką tik sukurtą funkciją,
# suskaičiuokite kiek vynų aprašymų (description) naudoja žodį 'tropical'
```

```python
reviews['is_tropical'] = reviews['description'].map(is_tropical)
```
